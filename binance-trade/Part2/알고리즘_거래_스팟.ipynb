## 롱 가격/거래량 거래 전략

높은 가격 상승과 거래량의 하락 클러스터를 이용

```python
data
```

```python
data["position"] = 1
data
```

매도하고 중립 포지션으로 전환, 만약 가격 상승이 매우 높고 거래량이 하락이라면

```python
# 매우 높은 수익률에 대한 임계값 받기(90 백분위 수 이상)
return_thresh = np.percentile(data.returns.dropna(), 90)
return_thresh
```

```python
cond1 = data.returns >= return_thresh
cond1
```

```python
# 적절한 거래량 감소에 대한 거래량 임계값 얻기(5에서 20 사이의 백분위수)
volume_thresh = np.percentile(data.vol_ch.dropna(), [5, 20])
volume_thresh
```

```python
cond2 = data.vol_ch.between(volume_thresh[0], volume_thresh[1])
cond2
```

```python
data.loc[cond1 & cond2, "position"] = 0
data
```

```python
data.position.value_counts()
```

```python
data.loc[:, "position"].plot(figsize=(12, 8))
plt.show()
```

```python
data.loc["2019-06", 'position'].plot(figsize=(12, 8))
plt.show()
```

## 전략 백테스팅

```python
data
```

```python
data.position.shift(1)
```

```python
data["strategy"] = data.position.shift(1) * data["returns"]
data
```

```python
data.loc["2017-08-19"]
```

```python
data[["returns", "strategy"]].sum().apply(np.exp)
```

```python
# 시작값 1로 전략의 수익률의 정규화표
data["cstrategy"] = data["strategy"].cumsum().apply(np.exp)
```

```python
data
```

```python
data[["creturns", "cstrategy"]].plot(figsize=(12, 8), fontsize=12)
plt.show()
```

```python
tp_year = 24 * 365.25
tp_year
```

```python
ann_mean = data[["returns", "strategy"]].mean() * tp_year
ann_mean
```

```python
ann_std = data[["returns", "strategy"]].std() * np.sqrt(tp_year)
ann_std
```

```python
sharpe = (np.exp(ann_mean) - 1) / ann_std
sharpe
```

## 거래 비용

```python
data.position.value_counts()
```

```python
data.position.diff().fillna(0).abs()
```

```python
data["trades"] = data.position.diff().fillna(0).abs()
```

```python
data.trades.value_counts()
```

```python
commissions = 0.00075
commissions
```

```python
other = 0.0001
other
```

```python
# 전체 거래 비용의 비율
ptc = np.log(1 - commissions) + np.log(1 - other)
ptc
```

```python
data["strategy_net"] = data.strategy + data.trades * ptc
data
```

```python
data["cstrategy_net"] = data.strategy_net.cumsum().apply(np.exp)
```

```python
data
```

```python
data[["creturns", "cstrategy", "cstrategy_net"]].plot(figsize=(12, 8))
plt.show()
```
