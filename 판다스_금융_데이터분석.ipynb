## 수익의 정규성

```python
import pandas as pd
import numpy as np
pd.options.display.float_format = '{:.4f}'.format
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8')
```

```python
msft = pd.read_csv("msft.csv", index_col="Date", parse_dates=["Date"])
msft
```

```python
msft.log_ret.plot(kind="hist", figsize=(15, 8), bins=100, fontsize=15, density=True)
plt.xlabel("Daily Returns", fontsize=15)
plt.ylabel("Frequency", fontsize=15)
plt.title("Frequency Distribution of Returns", fontsize=20)
plt.show()
```

```python
mu = msft.log_ret.mean()
mu
```

```python
sigma = msft.log_ret.std()
sigma
```

```python
import scipy.stats as stats
```

**왜도(skew)**

분포의 비대칭도
* 정규분포 = 왜도 0
* 왼쪽으로 치우침 = 왜도 > 0
* 오른쪽으로 치우침 = 왜도 < 0

```python
stats.skew(msft.log_ret.dropna())
```

**첨도(kurtosis)**

확률분포의 뾰족한 정도
* 정규 분포 = 첨도(0), Pearson 첨도(3)
* 위로 뾰족함 = 첨도(>0), Pearson 첨도(>3)
* 아래로 뾰족함 = 첨도(<0), Pearson 첨도(<3)

```python
stats.kurtosis(msft.log_ret.dropna(), fisher = True)
```

```python
x = np.linspace(msft.log_ret.min(), msft.log_ret.max(), 10000)
x
```

```python
y = stats.norm.pdf(x, loc=mu, scale=sigma)
y
```

```python
plt.figure(figsize=(20, 8))
plt.hist(msft.log_ret, bins = 500, density=True, label="Frequency Distribution of daily Returns (MSFT)")
plt.plot(x, y, linewidth = 3, color = 'red', label='Normal Distribution')
plt.title('Normal Distribution', fontsize=20)
plt.xlabel('Daily Returns', fontsize=15)
plt.ylabel('pdf', fontsize=15)
plt.legend(fontsize=15)
plt.show()
```

**MSFT 의 정규화된 수익을 테스트(2014년부터 2021년까지의 샘플)**

**통계상으로 증명할 수 없으므로 가설을 만들고 참인지 거짓인지 확인**
*  가설0 : MSFT 수익률은 정규 분포를 따른다

```python
z_stat, p_value = stats.normaltest(msft.log_ret.dropna())
```

```python
z_stat
```

```python
p_value
```

```python
round(p_value, 10)
```

* MSFT 수익률 값이 정규분포를 따른다는 가설은 0퍼센트의 확률이다
* 극단적인 상황과 결과는 평균 분산 분석에 반영되지 않으므로, 수익률의 표준 편차는 실제 위험을 과소 평가한다.
